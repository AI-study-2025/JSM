```python
import torch
import torch.nn as nn
from torchvision import models
from pprint import pprint

resnet18 = models.resnet18(pretrained=True)

pprint(resnet18)
```

```python
pprint(list(resnet18.children()))
```

```python
pprint(list(resnet18.named_children()))
# 애는 이름이랑 실제 모듈이 (이름, 모듈) 이렇게 저장되어있다.
```

```python
pprint(list(resnet18.modules()))
```

```python
# 특정 블럭을 꺼내는 방법

layer4 = resnet18.layer4
print(list(layer4.children()))


print(f'이중 첫번쨰 블럭을 꺼내는 방법 {list(layer4.children())[0]}')
print(f'이중 두번쨰 블럭을 꺼내는 방법 {list(layer4.children())[1]}')
```

```python
layers = list(resnet18.children())
for idx, module in enumerate(layers):
    print(f'{idx}번쨰를 나타내는 모듈 :  {module}')
```

```python
# 그러면 FC Layer를 적용하기 전에 feature map을 추출하고 싶다면 


backbone_layers = list(resnet18.children())
backbone = nn.Sequential(*backbone_layers[:8])

# print(list(backbone.named_children()))

data = torch.randn(1,3,224,224)
out_feature = backbone(data)
print(out_feature.size())
```

정리해보면 우리는 크게 4가지 방법으로 모델의 구조를 파악할수 있습니다. <br>

1) print(model)
2) print(list(model.children()))
3) print(list(model.named_children()))
4) print(list(model.modules()))

이를 통해서 우리가 원하는 모듈을 추출하거나 서브 모듈을 만들어서 재사용할수 있습니다.
